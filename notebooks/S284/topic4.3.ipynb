## 3.1 The birth of the Milky Way

### 3.1.1 The outskirts and neighbourhood of the Milky Way

Some **globular clusters** were formed within the Galaxy at a much earlier stage in its history, while others formed in smaller galaxies that accreted onto the Milky Way at various points in its history.

There are many **dwarf galaxies** currently in orbit around the much larger Milky Way. The two most prominent are the Large Magellanic Cloud (LMC) and the Small Magellanic Cloud (SMC).

Strips of material from dwarf galaxies eventually mix into the Milky Way via an accretion process that is thought to have been an important part of the Galaxys formation process.

Gravitiy is the dirving force behind this accretion process, but this is also true at scales much larger than the Milky Way itself. 

### 3.1.2 Models for how the Milky Way formed

It is believed that collapse from a single large cloud was responsible for the formation of a **protogalaxy** that eventually became the Millky Way, with accretion of inflowing gas and smaller galaxies gradually building up its size.

### 3.1.3 Examining the formation of the Milky Way

Like in the star formation process, **gravitational collapse** is the driver for galaxy matter to clump together, **rotation** leads to the formation of flattened discs, and **inflowing**, accreting gas also strongly influences formation process of galaxies like the Milky Way.

It is also thought that fragmentation within the collapsing protogalaxy was repsonsible for the **formation of globular clusters**.

An important difference between star and galaxy formation is the gravitational influence of **dark matter** in which the gas is thought to be embedded. 

Another difference is the **much longer timescales** on which the galaxy formation process operates.



## 3.3 Gravity and energy in galaxy formation

Gravity is the fundamental force controlling the process of making a galaxy. However, while molecular clouds collapse under the influence of their own gravity, it is dark matter what is thought to work at galactic scales.

### 3.3.1 Gravitational collapse on galactic scales

The Jeans mass is a less useful concept in galaxy formation. Unlike the case for a forming star, which is self-gravitating, a protogalactic gas cloud is thought to be embedded in a dark matter halo, whose gravity dominates the collapse process.

In the absence of any opposing force, gas will collapse inwards undter the influence of gravity - this is known as _free fall_. The **free-call timescale** for a gas cloud to collapse completely (to zero size) is given by

$$t = \sqrt{\frac{3 \pi}{32G\rho}}$$

where $\rho$ is the initial density of the cloud. In the case of star formation, we only need to consider the density of molecular gas.

The free-fall timescale is the _shortest possible_ time for a gas cloud to collapse when there is no force opposing gravity. Neither for stars nor galaxies, this reflects the true collapse time. We have seen the outward pressure of the heated up gas counteracts the inward force of gravity, slowing down (and sometimes stopping) the collapse.

### 3.3.2 Hydrostatic equilibrium from stars to galaxies

The timescale for gas to collapse is slowed down or halted as the cloud reaches hydristatic equilibrium. For a cloud to be stable, the forces acting on each point of the cloud need to balance out.

Using this basic physics principle, we can calculate the strength of the gravitational force needed to balance out the hot gas pressure measured (via X-ray spectra) in galaxy clusters. This allows us to sum up the total mass in galaxy clusters, with the answer that around **ten times more mass** is present than is comprised by the radiating material. This is one of the strongest pieces of evidence for **dark matter**.

### 3.3.3 Energy balance and heating in galaxy formation

The total **gravitational potential energy** released in forming a star, a galaxy or a galaxy cluster, through collapse from a larger gas cloud, can be approximated as

$$
E \sim \frac{GM_{tot}M_{gas}}{R}
$$

The released **gravitational energy** is converted to heat, causing the gas temeprature to rise:

$$
T \sim \frac{GMm}{kR}
$$

where $M$ is here the total mass of the system and we can make the simplifying assumption that the gas is all ionosied hydrogen to $m=0.5m_H$.

But most of the mass in the Milky Way’s ISM is very much cooler than this. The reason its temperature comes out much too high in our estimate is that we have so far ignored a very important process that alters the energy balance as stars and galaxies form: radiation!

### 3.3.4 Radiation and gas cooling to form galaxy discs

The gas that forms stars, galaxies, and clusters of galaxies emits radiation across a range of frequencies, depending on temperature. If that radiation can escape from the gas cloud it will carry energy away, and so the gas will cool.

The luminosity of a star or galaxy, which is measured in units of W (i.e. energy per unit time) is a measure of brightness, but it is also a rate at which energy is being lost from the object that’s radiating. In the case of a main sequence star, the energy lost to radiation is entirely compensated for by energy generated in the star’s core by nuclear burning.

Much of the gas in the ISM of the Milky Way is very cold, and yet the gas in between galaxies is very hot, emitting X-rays. This is a consequence of the large gravitational potential energy released as gas falls into a massive dark-matter halo.

If we know how luminous the cluster gas is, we can estimate how long it will take to cool by losing energy through radiation and then trigger galaxy formation:

$$
t = \frac{E_{th}}{L}
$$

where $E_{th}$ is the region total thermal energy and $L$ its luminosity. The thermal energy is defined as:

$$
E_{th} = \frac{3}{2}NkT
$$

with $N$ being the total number of particles, k the Boltzman constant, and T the gas temperature.

Most of the gas heated up in forming a typical galaxy cluster has not had time to lose enough radiation and cool down significantly.

Gas in the central regions of clusters (and smaller galaxy groups like the neighbourhood of the Milky Way) does cool, by emitting radiation at a range of wavelengths (including strong emission lines) that is sufficiently bright that it carries away a lot of energy (a process known as radiative cooling). Therefore the gas towards the centre of galaxy clusters is able to cool quickly enough to explain the existence of cold gas in galaxies. 

The basic picture is that gas that is initially heated to high temperatures by gravitational collapse later cools and forms the discs of molecular and atomic gas we observe in galaxies like the Milky Way.

Finally, when molecular clouds within those cold gas discs collapse to form stars, both heating and cooling through radiation are important, until nuclear burning leads to a balance between energy produced and radiated away.

### 3.3.5 The effect of galaxy mergers

As well as causing clouds of gas to collapse, gravity also leads to mergers between galaxies both during the process of formation and after they have already collapsed to a disc and formed many stars.

**Major mergers** are collisions between two galaxies of similar mass. In this situation, both galaxies are substantially altered by the process, merging to form a galaxy that may be quite different from either of the original two components. Spiral structure is usually lost in a major merger, unless both galaxies are gas-rich. In this case a disc can reform, or form in the first place if the gas-rich progenitors did not already have discs. In a major merger between two gas-poor galaxies, spiral structure cannot form; if one galaxy is gas-rich and the other gas-poor, the resulting structure is uncertain.

**Minor mergers** are collisions where one galaxy is much more massive than the other. In this situation, the smaller galaxy will typically suffer more catastrophic consequences than the larger, so that the resulting galaxy retains the structure of the more massive galaxy, irrespective of whether it is gas-rich or gas-poor.


## 3.4 The dynamic cosmic web

### 3.4.1 The hierarchical model of galaxy formation

Modern understanding of how galaxies form and grow within the cosmic web has been built up through a combination of increasingly deep galaxy surveys together with progressively more powerful computer models and simulations.

The cosmic web we see today is thought to have grown from small variations in the density of gas and dark matter soon after the big bang, which were then pulled together by gravity. Galaxy formation simulations take this initial distribution of matter and then apply a set of rules to follow how matter clumps together and forms galaxies. These rules are usually based on a combination of firmly-understood physics together with some less certain assumptions to be tested.

It is important to recognise that this clumping together of matter takes place within a Universe that is expanding (see the discussion of redshift in Topic 2 Part 3) – the simulations must take this into account. But in the denser parts of the early cosmic web, gravity is sufficient to overcome this expansion, allowing matter (both dark matter and gas) to collapse to form larger structures.

### 3.4.2 Computer models of galaxy formation

There are two main computer modelling approaches to investigating hierarchical galaxy formation: hydrodynamic simulations and semi-analytic models. Both methods start from initial conditions in the early Universe and then gradually step forward in time to trace how matter evolves to form galaxies. They take different approaches to the rules that are applied at each time step.

##### Hydrodynamic simulations

Hydrodynamic simulations treat gas and dark matter as fluids. The region of space being modelled is a 3D box, which is split into either a grid of small regions or parcels of material whose properties, such as density and temperature, are tracked. These properties evolve according to well-understood physical laws.

A major challenge is the enormous range of size scales that are important for getting the physics right. It is infeasible to carry out a single simulation that traces the formation of galaxy clusters at the largest scales and also includes enough detail to form individual stars, which are a factor of more than 1013 times smaller.

State-of-the-art cosmological simulations now use sophisticated ways to adapt to the evolving structure. One method involves tracking a grid of regions that move along with the gas as it evolves, with the grid expanding and contracting so that each region continues to contain the same amount of gas as its density and other physical properties change. Another widely-used approach is to simulate a large volume of the Universe at lower resolution, and then, based on the results that emerge, do new zoom-in simulations at higher resolution of particular regions of interest.

##### Semi-analytic models

Semi-analytic modelling is a faster method than hydrodynamic analysis and one that requires less computer processing time, but does not attempt to follow the physics of how gas evolves in detail. Instead, the clumps of matter within a merger tree (corresponding to dark-matter halos), are traced as they merge together. Simpler rules or prescriptions are then adopted to determine how the gas evolves and how stars form within the halos as they merge.

### 3.4.3 Exploring the forming Universe

**Can you see any difference between the distributions of gas and dark matter?**

Not all of the dark-matter halos (clumps) contain gas – there are many small dark-matter clumps that don’t show up as corresponding clumps in the gas map. Because the gravitational pull of these clumps is small, gas can escape their gravity easily and tends to get blown out by small amounts of star formation; the gas is thus only retained in larger clumps. It is thought that many such dark-matter clumps (i.e. small ones that don’t retain gas) may exist in the real Universe
